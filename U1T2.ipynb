#Analyzing Execution Times

```python
import time
import random
import matplotlib.pyplot as plt

def sum_values(values):
    total = 0
    for value in values:
        total += value
    return total

def gen_input(length):
    return [random.randint(-1000, 1000) for _ in range(length)]

def plot_times(times):
    plt.plot(times)
    plt.xlabel('Size')
    plt.ylabel('Runtime')
    plt.title('Runtime vs Size')
    plt.show()

times = []

for length in range(1, 501):
    values = gen_input(length)
    start = time.time()
    sum_values(values)
    end = time.time()
    execution_time = end - start
    times.append(execution_time)

plot_times(times)
```

#Complexity of Function Calls

```python
def sum_values(values):
    total = 0            # 1
    for value in values: # N
        total += value   # N
    return total         # 1

def num_values(values):
    total = 0            # 1
    for _ in values:     # N
        total += 1       # N
    return total         # 1

def average(values):
    value_sum = sum_values(values)  # N
    num_values = num_values(values) # N
    return value_sum / num_values   # 1
```

#Fundamental Operations of a Computer

```python
import matplotlib.pyplot as plt

def plot_costs(times):
    plt.plot(times, label='append cost')
    plt.ylabel('cost')
    plt.xlabel('N')
    plt.show()

def append_cost(array_length, list_length):
    if array_length == list_length:
        return array_length
    return 1

def append_N_list_cost(N):
    array_length = 1  # initially the array will have length 1
    list_length = 0   # initially the list has 0 elements
    total_cost = 0    # this variable will keep track of the total cost
    for i in range(N):
        cost = append_cost(array_length, list_length)
        total_cost += cost
        # update the array and list lengths
        if array_length == list_length:
            array_length *= 2
        list_length += 1
    return total_cost

# Initialize an empty list named costs
costs = []

# Use a for loop over range(5000) using a variable named N
for N in range(5000):
    # Call append_N_list_cost() by passing the current value of N as an argument
    cost = append_N_list_cost(N)
    # Append the result to the costs list
    costs.append(cost)

# Use the provided plot_costs() function to plot the costs list
plot_costs(costs)
```

#List Complexities

```python
import time

def add_with_append(N):
    values = []
    for i in range(N):
        values.append(i)
    return values

def add_with_insert(N):
    values = []
    for i in range(N):
        values.insert(0, i)
    return values

# Time the execution of add_with_append(50000)
start = time.time()
add_with_append(50000)
end = time.time()
time_append = end - start

# Time the execution of add_with_insert(50000)
start = time.time()
add_with_insert(50000)
end = time.time()
time_insert = end - start

# Print both times
print(f"Time taken with append: {time_append} seconds")
print(f"Time taken with insert: {time_insert} seconds")
```

#Arithmetic Operations

```python
import random
import time
import matplotlib.pyplot as plt

def generate_list(k):
    items = []
    count = 1
    while len(items) < k:
        for _ in range(count):
            num1 = random.randint(10**(count-1), 10**count-1)
            num2 = random.randint(10**(count-1), 10**count-1)
            items.append((num1, num2))
            count += 1
            if len(items) == 5000:
                break
    return items

pairs = generate_list(5000)

def plot_times(times):
    plt.plot(times)
    plt.ylabel('multiplication runtime')
    plt.xlabel('number of digits')
    plt.yscale('log')
    plt.yticks([])
    plt.show()

# Print the first few pairs to verify the data
print(pairs[:5])

# Initialize an empty list named times
times = []

# Iterate over all pairs in the pairs list and measure the time for multiplication
for pair in pairs:
    start = time.time()
    result = pair[0] * pair[1]
    end = time.time()
    elapsed_time = end - start
    times.append(elapsed_time)

# Use the provided plot_times() function to plot the times list
plot_times(times)
```

#String Concatenation

```python
import random
import string
import time

def generate_list(k):
    result = []
    for _ in range(k):
        string_length = 10
        random_string = ''.join(random.choices(string.ascii_letters + string.digits, k=string_length))
        result.append(random_string)
    return result

def concat_with_add(word_list):
    concat = ''
    for word in word_list:
        concat += word
    return concat

def concat_with_join(word_list):
    return ''.join(word_list)

k = 100000  # Define the number of positions in the list
random_strings = generate_list(k)

# Print the first 5 elements of random_strings to verify the data
print(random_strings[:5])

# Measure the execution time of concat_with_add() over random_strings
start = time.time()
concat_with_add(random_strings)
end = time.time()
time_add = end - start

# Measure the execution time of concat_with_join() over random_strings
start = time.time()
concat_with_join(random_strings)
end = time.time()
time_join = end - start

# Print both times
print(f"Time taken with concat_with_add: {time_add} seconds")
print(f"Time taken with concat_with_join: {time_join} seconds")
```
